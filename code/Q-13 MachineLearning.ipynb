```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("C:\\Users\\rcc\\Desktop\\USA_Housing.csv")
```

```python
data=data.drop(['Address'],axis=1)
```

```python
data.info()
```

```python
data.describe()
```

```python
data.duplicated().sum()
```

```python
data.clip(data.quantile(0.1),data.quantile(0.9),inplace=True,axis=1)
```

```python
plt.boxplot(data)
plt.show()
```

```python
from sklearn.preprocessing import Normalizer,MinMaxScaler,StandardScaler,RobustScaler
```

```python
Norm=Normalizer()
MK=Norm.fit_transform(data)
data1=pd.DataFrame(MK,columns=data.columns)
data1
```

```python
Min=MinMaxScaler()
MK1=Min.fit_transform(data)
data2=pd.DataFrame(MK1,columns=data.columns)
data2
```

```python
Stan=StandardScaler()
MK2=Stan.fit_transform(data)
data3=pd.DataFrame(MK2,columns=data.columns)
data3
```

```python
Rob=RobustScaler()
MK3=Rob.fit_transform(data)
data4=pd.DataFrame(MK3,columns=data.columns)
data4
```
